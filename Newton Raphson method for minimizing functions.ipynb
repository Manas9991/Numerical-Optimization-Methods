##Inputs: 

```
#Define objective function here:
def f(x):
   return (x*x)+(54/x)

#Enter bounds:

xmin = 0
xmax = 5

#Enter resolution:

e = 0.01
```

##Solve:

```
#to visualize convergence
import matplotlib.pyplot as plt
%matplotlib inline
pts = [xmax] #storing values of x to plot later

#solution:
err = xmax-xmin
dx = 0.01*xmax

while ((err>=e)and(xmin<=xmax)):
  suma = (xmax+dx)
  difa = (xmax-dx)
  xnew = xmax - ( ( (f(suma)-f(difa)) / (2*dx) ) / ( (f(suma)-(2*f(xmax))+f(difa)) /(dx*dx) )  )
  pts.append(xnew)
  err = xmax-xnew
  xmax = xnew
```

##Results:

```
print("Point of minima: ", round(xmax, 2))
print("Optimum value: ", round(f(xmax), 2))
```

##Plots:

```
x = []
i = 1
while (i<= 5):
  x.append(i)
  i += 0.01
fx = [f(X) for X in x]
low = [f(xmax)]*len(x)

plt.plot(x, fx, color='black')
plt.plot(x, low, color='orange')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```
plt.plot(range(len(pts)), pts, color='red')
plt.title('Convergence of x value vs iterations')
plt.xlabel('Iterations')
plt.ylabel('x')
plt.show()
pts = [(int(i*100))/100 for i in pts] #another way of round(x, 2)
print("x values: ", pts)
```
